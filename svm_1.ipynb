```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# creating datapoints
from sklearn.datasets import make_classification
```

```python
X,y=make_classification(n_samples=1000,n_features=2,n_classes=2,n_clusters_per_class=2,n_redundant=0,random_state=50)
```

```python
X
```

```python
y
```

```python
pd.DataFrame(X)[0]
```

```python
sns.scatterplot(x=pd.DataFrame(X)[0],y=pd.DataFrame(X)[1], hue=y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
from sklearn.svm import SVC
```

```python
svc=SVC(kernel='linear')
```

```python
svc.fit(X_train,y_train)
```

```python
ypred=svc.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,ypred))
print(confusion_matrix(y_test,ypred))
```

```python
poly=SVC(kernel='poly')
poly.fit(X_train,y_train)
ypred1=poly.predict(X_test)
print(classification_report(y_test,ypred1))
print(confusion_matrix(y_test,ypred1))
```

```python
rbf=SVC(kernel='rbf')
rbf.fit(X_train,y_train)
ypred2=rbf.predict(X_test)
print(classification_report(y_test,ypred2))
print(confusion_matrix(y_test,ypred2))
```

```python
sigmoid=SVC(kernel='sigmoid')
sigmoid.fit(X_train,y_train)
ypred3=sigmoid.predict(X_test)
print(classification_report(y_test,ypred3))
print(confusion_matrix(y_test,ypred3))
```

```python
from sklearn.model_selection import GridSearchCV

param_grid={'C':[.1,1,10,100,100],

            'gamma':[10,20,1,.1,.01,.001,.0001],
            'kernel':['rbf']}
```

```python
grid=GridSearchCV(SVC(),param_grid=param_grid,cv=5,verbose=3,refit=True)
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
ypred4=grid.predict(X_test)
print(classification_report(y_test,ypred4))
print(confusion_matrix(y_test,ypred4))
```
